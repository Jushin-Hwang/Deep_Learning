# 6) Model 불러오기

```python
# Base 모델을 로드 (Abalone_CNN.h5)
from tensorflow.keras.models import load_model

model_file = './Abalone_CNN.h5'

loaded_model = load_model(model_file)
```

```python
# 기존 모델의 층을 동결 (학습되지 않도록 설정)
loaded_model.trainable = False
```

```python
from tensorflow.keras import layers, models
from keras.layers import Conv1D, Dense
from keras.models import Sequential

def remove_last_layers(model, num_layers_to_remove):
    # 모델의 레이어를 하나씩 슬라이스해서 앞의 레이어만 가져옴
    model_layers = model.layers[:-num_layers_to_remove]  # 마지막 레이어부터 지정된 갯수만큼 제외
    new_model = models.Sequential(model_layers)  # 새로운 모델에 해당 레이어들만 추가
    return new_model

# model에서 특정 레이어만을 추출하는 함수
def my_layers(model, start_index, end_index) :
    model_layers = model.layers[start_index : end_index + 1]
    return model_layers

# 기존 모델에서 레이어를 제거한 후 남은 레이어들을 새로운 Sequential 모델에 추가
def add_layers_to_new_model(new_model, layers_to_add):
    for layer in layers_to_add:
        new_model.add(layer)

new_model = Sequential()

# 새로운 입력층 추가
new_model.add(Conv1D(filters=32, kernel_size=2, activation='relu', input_shape=(278, 1)))

# 입력층 이후부터 dense_34 전까지의 layers 추가
layers = my_layers(loaded_model, 1, 2)
add_layers_to_new_model(new_model, layers)

# dense_34가 128의 입력을 받도록 설정되어있다는 오류 메세지 발생.
# 128의 출력을 가지는 Dense를 추가했음
new_model.add(Dense(128, activation = 'relu'))

# 나머지 레이어를 new_model에 추가
layers = my_layers(loaded_model, 3, 5)
add_layers_to_new_model(new_model, layers)

# 마지막층 1개의 레이어 제거
new_model = remove_last_layers(new_model, 1)

# 새로운 출력층 추가
new_model.add(Dense(32, activation='relu'))
new_model.add(Dense(16, activation='relu'))
new_model.add(Dense(19, activation='softmax'))  # 예: 19개의 클래스
new_model.compile(optimizer='adam', metrics=['accuracy'], loss='categorical_crossentropy')

# 모델 요약 출력
new_model.summary()
```

```python
import pandas as pd
display_layers = [(layer, layer.name, layer.trainable) for layer in new_model.layers]
pd.DataFrame(display_layers, columns=['Layer Type', 'Layer Name', 'Layer Trainable'])
```

기존 모델 6개의 layer중 입력층과 출력층을 제외한 4개의 layer들이 잘 전이되었음을 확인할 수 있다.

# 7) 학습, 예측

```python
history = new_model.fit(X_train, Y_train, epochs = 20, batch_size = 32, validation_split = 0.1, shuffle = True)

new_model.summary()
```

```python
new_model.evaluate(X_test, Y_test)
```

# 8) Fine Tunning

```python
for layer in new_model.layers:
    layer.trainable = True

new_model.compile(optimizer='adam', metrics=['accuracy'], loss='categorical_crossentropy')
```

```python
import pandas as pd
display_layers = [(layer, layer.name, layer.trainable) for layer in new_model.layers]
pd.DataFrame(display_layers, columns=['Layer Type', 'Layer Name', 'Layer Trainable'])
```

```python
import numpy as np

history2 = new_model.fit(X_train, Y_train, epochs = 20, batch_size = 32, validation_split = 0.1, shuffle = True)

Y_pred = new_model.predict(X_test)

Y_pred_class = np.argmax(Y_pred, axis = -1)

new_model.summary()
```

```python
new_model.evaluate(X_test, Y_test)
```

# 10) 시각화

```python
import matplotlib.pyplot as plt

plt.plot(history2.history["loss"], label = "train")
plt.plot(history2.history["val_loss"], label = "test")
plt.legend()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

Y_test_index = np.argmax(Y_test, axis=1)

# 혼동 행렬
cm = confusion_matrix(Y_test_index, Y_pred_class)

plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()

# 성능 보고서
print(classification_report(Y_test_index, Y_pred_class))
```
